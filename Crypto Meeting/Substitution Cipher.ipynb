```python
import csv
reader = csv.reader(open("freqs.csv", "r"))
real_freqs = {float(v): k for k, v in reader}
print(real_freqs)
```

```python
from collections import Counter
text = ''
freqs = Counter(text)
result_freqs = {k: v / len(text) for k, v in freqs.items()}
print(result_freqs)
```

```python
mapping = {}
for k in result_freqs:
    mapping[k] = real_freqs[result_freqs[k]]

print(mapping)
```

```python
decrypted = ''.join(mapping[char] for char in text)
print(decrypted)
```
